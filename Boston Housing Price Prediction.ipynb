```python
from sklearn.model_selection import StratifiedShuffleSplit
shuffler=StratifiedShuffleSplit(n_splits=1, test_size=0.15, random_state=9)
```

```python
for train,test in shuffler.split(df, df['CHAS']):
    stra_train_set=df.iloc[train]
    stra_test_set = df.iloc[test]
```

## 

```python
X_train= stra_train_set.iloc[:,:-1].values
y_train= stra_train_set.iloc[:,-1].values
```

```python
X_test= stra_test_set.iloc[:,:-1].values
y_test= stra_test_set.iloc[:,-1].values
```

```python
from sklearn.linear_model import LinearRegression
regressor= LinearRegression()
regressor.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score, clf
predictions= regressor.predict(X_test)
```

```python
print(r2_score(y_test,predictions))
```
